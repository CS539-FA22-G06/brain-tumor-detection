```python
from zipfile import ZipFile
import os
import tqdm
from google.colab import drive
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
import json
```

```python
# Connecting to colab
drive.mount("/content/drive")
data_path = "/content/drive/Shareddrives/CS539 Group 6/data/kaggle_br35h.zip"
```

```python
imgzip = ZipFile(data_path)
inflist = imgzip.namelist()[1:]

images = []

#Create X_train array for BR35h dataset

for f in inflist:
  #Grab jpegs of NO TUMOR
  if(".jpg" in f and "no" in f):
    ifile = imgzip.open(f)
    img = Image.open(ifile)
    scan = np.asarray(img)

    #Reshape images into (150,150,3)
    if(len(scan.shape)<3):
      scan = np.stack((img,)*3, axis=-1)
    img = cv2.resize(scan,(150, 150))
    label = "no_tumor"
    
    images.append((label,img))
      
```

```python
print(len(images))
```

```python
data = [[x[0]] + x[1].flatten().tolist() for x in images]
df = pd.DataFrame(data)
#df.to_csv(f"/content/drive/Shareddrives/CS539 Group 6/data/br35h.csv")
```
