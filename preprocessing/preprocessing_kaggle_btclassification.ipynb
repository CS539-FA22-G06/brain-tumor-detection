```python
from zipfile import ZipFile
import cv2
import matplotlib.pyplot as plt
from io import BytesIO
import numpy as np
from google.colab import drive
import pandas as pd
from PIL import Image
```

```python
shapes = set()
names = set()
image_size = 150
images = []
data_path = "/content/drive/Shareddrives/CS539 Group 6/data/kaggle_btclassification.zip"
with ZipFile(f"{data_path}") as zf:
    for name in zf.namelist():
        # names.add(name.split('.')[-1])
        group, label = tuple(name.split('/')[0:-1])
        # continue
        f = zf.read(name)
        img = np.asarray(Image.open(BytesIO(f)))
        img = cv2.resize(img, (image_size, image_size))
        images.append((label, img))
        # print(img)
        
        # plt.imshow(img)
        # break

# print(shapes)
# print(sorted(names))
```

```python
data = [[x[0]] + x[1].flatten().tolist() for x in images]
df = pd.DataFrame(data)
df.to_csv(f"/content/drive/Shareddrives/CS539 Group 6/data/kaggle_btclassification_images.csv")
```
